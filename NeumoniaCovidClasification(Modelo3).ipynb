```python
from itertools import chain
from matplotlib import pyplot as plt

unique_dimensions = set()
unique_modes = set()

for subdir in chain(os.listdir(train_dir), os.listdir(test_dir), os.listdir(val_dir)):
    subdir_path = os.path.join(train_dir, subdir)
    for filename in os.listdir(subdir_path):
        file_path = os.path.join(subdir_path, filename)
        image = Image.open(file_path)
        unique_dimensions.add(image.size[0])
        unique_dimensions.add(image.size[1])
        unique_modes.add(image.mode)

print("unique modes:", unique_modes)
print("unique dimensions:", unique_dimensions)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale=1./255)
valid_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(299, 299),
        color_mode='grayscale',
        class_mode='binary')

valid_generator = valid_datagen.flow_from_directory(
        val_dir,
        target_size=(299, 299),
        color_mode='grayscale',
        class_mode='binary')

test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(299, 299),
        color_mode='grayscale',
        class_mode='binary',
        shuffle=False)
```

Red neuronal

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, BatchNormalization, Activation, Dropout
from keras.layers import Conv2D, MaxPooling2D
from keras import backend
```

```python
def onelayercnn():
    backend.clear_session()
    cnn1 = Sequential()
    cnn1.add(Conv2D(38, kernel_size=(3, 3), activation='relu', input_shape=(299, 299, 1)))
    cnn1.add(MaxPooling2D(pool_size=(2, 2)))
    cnn1.add(Flatten())
    cnn1.add(Dense(150, activation='relu'))
    cnn1.add(Dense(1, activation='sigmoid'))
    cnn1.summary()
    return cnn1
```

```python
cnn1 = onelayercnn()
```

```python
cnn1.compile(loss='binary_crossentropy',
              optimizer="Adam",
              metrics=['accuracy'])
```

```python

EPOCHS = 10
history = cnn1.fit(
      train_generator,
      validation_data=valid_generator,
      epochs=EPOCHS,
      verbose=1)
```

```python
y_true = test_generator.labels
y_pred = cnn1.predict(test_generator)
y_pred = np.round(y_pred)
print(classification_report(y_true, y_pred,digits=4))
```

```python
plt.figure(figsize=(8,6))
plt.title("Loss Curves")
plt.plot(history.history["loss"], label="loss")
plt.plot(history.history["val_loss"], label="val_loss")
plt.ylim([0,1.5])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title("Accuracy Curves")
plt.plot(history.history["accuracy"], label="accuracy")
plt.plot(history.history["val_accuracy"], label="val_accuracy")
plt.ylim([0,1.5])
plt.legend()
plt.show()
```
