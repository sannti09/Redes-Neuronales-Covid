Funcion para aplicar mascara de las imagenes

```python
def maskimage(path):
  img = cv2.imread(path,0)
  crop_img = img[25:281, 28:284].copy()

  path2 = path.replace("images", "masks")
  covidmaskaux = cv2.imread(path2,0)

  result = cv2.bitwise_and(crop_img,crop_img, mask= covidmaskaux)
  return result
```

Entrenamiento del modelo

```python
train_dir='/content/drive/MyDrive/Tecnicas/Radiografias/train'
val_dir='/content/drive/MyDrive/Tecnicas/Radiografias/val'
test_dir= '/content/drive/MyDrive/Tecnicas/Radiografias/test'
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale=1./255)
valid_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(256, 256),
        color_mode='grayscale',
        class_mode='binary')

valid_generator = valid_datagen.flow_from_directory(
        val_dir,
        target_size=(256, 256),
        color_mode='grayscale',
        class_mode='binary')

test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(256, 256),
        color_mode='grayscale',
        class_mode='binary',
        shuffle=False)
```

Red neuronal

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, BatchNormalization, Activation, Dropout
from keras.layers import Conv2D, MaxPooling2D
from keras import backend
```

```python
def onelayercnn():
    backend.clear_session()
    cnnmask = Sequential()
    cnnmask.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(256, 256, 1)))
    cnnmask.add(MaxPooling2D(pool_size=(2, 2)))
    cnnmask.add(Dropout(0.2))
    cnnmask.add(Flatten())
    cnnmask.add(Dense(128, activation='relu'))
    cnnmask.add(Dense(1, activation='sigmoid'))
    cnnmask.summary()
    return cnnmask
```

```python
cnnmask = onelayercnn()
```

```python
cnnmask.compile(loss='binary_crossentropy',
              optimizer="Adam",
              metrics=['accuracy'])
```

```python

EPOCHS = 50
history = cnnmask.fit(
      train_generator,
      validation_data=valid_generator,
      epochs=EPOCHS,
      verbose=1)
```

```python
y_true = test_generator.labels
y_pred = cnnmask.predict(test_generator)
y_pred = np.round(y_pred)
print(classification_report(y_true, y_pred,digits=4))
```

```python
plt.figure(figsize=(8,6))
plt.title("Loss Curves")
plt.plot(history.history["loss"], label="loss")
plt.plot(history.history["val_loss"], label="val_loss")
plt.ylim([0,1.5])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title("Accuracy Curves")
plt.plot(history.history["accuracy"], label="accuracy")
plt.plot(history.history["val_accuracy"], label="val_accuracy")
plt.ylim([0,1.5])
plt.legend()
plt.show()
```
